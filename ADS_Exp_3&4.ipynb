# Demographic filtering Recommender system


Demographic Filtering -

Before getting started with this -

    we need a metric to score or rate movie
    Calculate the score for every movie
    Sort the scores and recommend the best rated movie to the users.

We can use the average ratings of the movie as the score but using this won't be fair enough since a movie with 8.9 average rating and only 3 votes cannot be considered better than the movie with 7.8 as as average rating but 40 votes. So, I'll be using IMDB's weighted rating (wr) which is given as :-

where,

    v is the number of votes for the movie;
    m is the minimum votes required to be listed in the chart;
    R is the average rating of the movie; And
    C is the mean vote across the whole report

We already have v(vote_count) and R (vote_average) and C can be calculated as


```python
C=df1['avg_rating'].mean()
C
```

```python
m= df1['vote_count'].quantile(0.9)
m
```

```python
q_movies = df1.copy().loc[df1['vote_count'] >= m]
q_movies.shape
```

```python
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['avg_rating']
    # Calculation based on the IMDB formula
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
```

```python
q_movies = q_movies.sort_values('score', ascending=False)
q_movies.head(10)
```

```python
df5=df5.drop('timestamp',axis=1)
df5.head()
```

```python
useritem = np.zeros([700, 10000])
for userid in range(len(df5['userId'])):
    try:
        useritem[df5['userId'][userid]-1][df5['movieId'][userid]-1]=df5['rating'][userid]
    except:
        pass
```

```python
moviecolumns=[]
for movieval in range(len(useritem[0])):
    moviecolumns.append("movie "+str(movieval))
```

```python
mdf=pd.DataFrame(useritem,columns=moviecolumns)
mdf.head()
```

```python
mdf['movie 0'][[0,1]]
```

```python
mdf.loc[0].mean()
```

```python
def standerdize(row):
    new_row=(row-row.mean())
    return new_row
```

```python
smdf=mdf.apply(standerdize,axis=1)
smdf.head()
```

```python
smdf=smdf[mdf>0]
smdf=smdf.fillna(0)
```

```python
smdf.head()
```

```python
import numpy as np
```

```python
from sklearn.metrics.pairwise import cosine_similarity
item_similarity=cosine_similarity(smdf)
```

```python
item_similarity
```

```python
usercol=["user"+str(x) for x in range(700)]
```

```python
its=pd.DataFrame(item_similarity,columns=usercol)
```

```python
its.shape
```

```python
def movie_rate_prediction(user,movie_name):
    temp=mdf[movie_name][list(its[user].sort_values(ascending=False).index[1:10])]
    return temp[temp>0].mean()
```
